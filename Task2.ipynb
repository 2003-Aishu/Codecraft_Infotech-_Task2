ðŸ”¹âœ… "To perform data cleaning and exploratory data analysis (EDA) on the Titanic dataset. This involves handling missing values, correcting data types, and exploring patterns and relationships between features (e.g., Age, Sex, Pclass) and the target variable Survived"

âœ… Step 1: Load Required Libraries & Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the dataset
df = pd.read_csv("C:/Users/Aishwarya/Downloads/train.csv")
```

âœ… Step 2: Data Cleaning (already done)

```python
# Fill missing Age with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill missing Embarked with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

âœ… Step 3: Basic Data Exploration

```python
# Basic info
df.info()

# Summary stats
df.describe()

# Check missing values
df.isnull().sum()
```

âœ… Step 4: Univariate Analysis (One Variable at a Time)

ðŸ”¹ Countplot - Survival

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()
```

ðŸ”¹ Distribution - Age

```python
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

ðŸ”¹ Countplot - Pclass

```python
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Count")
plt.show()
```

âœ… Step 5: Bivariate Analysis (Compare Two Variables)

ðŸ”¸ Survival by Sex

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival by Gender")
plt.show()
```

ðŸ”¸ Survival by Pclass

```python
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

ðŸ”¸ Age vs. Survival

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age vs. Survival")
plt.show()
```

âœ… Step 6: Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
